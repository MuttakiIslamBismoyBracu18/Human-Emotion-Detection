```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
import os
```

```python
def analyze_video_emotions(video_path, frame_interval=30):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print("Error: Unable to open video.")
        return []
    
    frame_count = 0
    emotions_over_time = []  # Store detected emotions per frame
    
    # Loop through the video frames
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print(f"Finished reading video at frame {frame_count}")
            break  # Exit if no more frames
        
        # Analyze every 'frame_interval' frame (to save processing time)
        if frame_count % frame_interval == 0:
            print(f"Processing frame {frame_count}...")  # Debug print
            
            try:
                # Step 4: Perform emotion detection on each frame
                result = DeepFace.analyze(frame, actions=['emotion'])
                emotions_over_time.append(result['emotion'])  # Append detected emotions
            except Exception as e:
                print(f"Error analyzing frame {frame_count}: {e}")
        
        frame_count += 1
    
    # Release the video capture object
    cap.release()
    
    if len(emotions_over_time) == 0:
        print("No emotions were detected in the video.")
    
    return emotions_over_time
```

```python
def plot_emotions(emotions_over_time):
    if len(emotions_over_time) == 0:
        print("No emotion data to plot.")
        return

    # Initialize data containers
    emotion_keys = emotions_over_time[0].keys()  # Emotion categories
    emotion_data = {key: [] for key in emotion_keys}  # To store emotion intensities

    # Organize emotion intensities by frames
    for emotions in emotions_over_time:
        for key in emotion_keys:
            emotion_data[key].append(emotions[key])

    # Step 6: Plot the emotion data
    plt.figure(figsize=(10, 6))
    
    for emotion, values in emotion_data.items():
        plt.plot(values, label=emotion)

    plt.title("Emotion Changes Over Time")
    plt.xlabel("Frame Index")
    plt.ylabel("Emotion Intensity")
    plt.legend(loc="upper right")
    plt.show()
```

```python
def main(video_path):
    emotions_over_time = analyze_video_emotions(video_path, frame_interval=30)
    plot_emotions(emotions_over_time)
```

```python
video_path = input("Please enter the full path to the video file: ")
```

```python
if os.path.isfile(video_path):
    main(video_path)
else:
    print("Error: Video file not found. Please provide a valid path.")
```
